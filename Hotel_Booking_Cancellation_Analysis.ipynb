# Hotel Booking Cancellation Analysis

## End-to-End Exploratory Data Analysis Project

This notebook analyzes hotel booking data to understand cancellation behavior
and provide actionable business insights for improving revenue and occupancy.


## Project Overview

Hotel booking cancellations negatively impact revenue, occupancy planning,
and operational efficiency. This project uses historical booking data to
identify patterns behind cancellations and support data-driven decision making.


## Business Context

Hotels rely heavily on advance bookings to manage pricing, staffing, and room
availability. When bookings are canceled—especially at the last moment—rooms
often remain unsold, leading to revenue loss.

City Hotels typically experience volatile demand due to business travel,
while Resort Hotels depend on seasonal leisure demand. Understanding why
customers cancel bookings is critical for improving profitability.


## Business Problem

Both City Hotels and Resort Hotels face high booking cancellation rates,
which result in:

- Revenue loss
- Under-utilized rooms
- Inefficient operational planning

The core problem is identifying the factors that drive cancellations and
using those insights to reduce revenue loss.


## Business Objectives

- Analyze patterns and trends in booking cancellations
- Identify key factors influencing cancellations
- Compare City Hotel and Resort Hotel behavior
- Provide actionable business recommendations


## Assumptions

1. No unusual global events affected the data (2015–2017)
2. The dataset represents real hotel booking behavior
3. Hotels are not currently using the proposed strategies
4. Booking cancellations significantly impact revenue


## Research Questions

1. What factors influence hotel booking cancellations?
2. Do City Hotels and Resort Hotels behave differently?
3. How do pricing (ADR) and lead time affect cancellations?
4. Are there seasonal trends in cancellations?


## Hypotheses

1. Higher ADR leads to more cancellations
2. Longer lead time increases cancellation probability
3. Booking channels influence cancellation behavior

## Cancellation Rate by Hotel Type


```python
plt.figure(figsize=(8,4))

ax1 = sns.countplot(
    x='hotel',
    hue='is_canceled',
    data=df,
    palette='Blues'
)

ax1.legend(bbox_to_anchor=(1, 1))
plt.title('Reservation status in different hotels', size=20)
plt.xlabel('Hotel')
plt.ylabel('Number of reservations')
plt.show()
```

### Insight
City Hotels experience significantly higher cancellation rates compared to
Resort Hotels, indicating more volatile demand.


```python
resort_hotel=df[df['hotel']=='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

In resort hotel the cancelation percentage is 27% that is less than the city hotels because the prices of resorts are higher than the city hotel we will prove it ahead

```python
resort_hotel=df[df['hotel']=='City Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
# Split data by hotel type
resort_hotel = df[df['hotel'] == 'Resort Hotel']
city_hotel   = df[df['hotel'] == 'City Hotel']

# Group by date and calculate mean ADR
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel   = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(20,8))
plt.title('Average daily rate in city and resort hotel')
plt.plot(resort_hotel.index,resort_hotel['adr'],label='Resort Hotel')
plt.plot(city_hotel.index,city_hotel['adr'],label='City Hotel')
plt.legend(fontsize=20)
plt.show()
```

```python
df['month']=df['reservation_status_date'].dt.month
plt.figure(figsize=(16,8))
ax1=sns.countplot(x='month',hue='is_canceled',data=df)
```

```python

sns.barplot(
    x='month',
    y='adr',
    data=df[df['is_canceled'] == 1]
         .groupby('month')[['adr']]
         .sum()
         .reset_index()
)

plt.show()
```

### Insight
Canceled bookings generally have higher ADR values, confirming that higher
prices increase cancellation likelihood.


```python
canceled_data=df[df['is_canceled']==1]
top_10_country=canceled_data['country'].value_counts()[:10]
plt.figure(figsize=(8,9))
plt.title('top 10 countries with reservation cancelledd')
plt.pie(top_10_country, autopct='%.2f',labels=top_10_country.index)
plt.show()
```

### Highest number of booking are coming through the online travel agent but the reservation that are cancelling also are online travel agent

```python
canceled_df_adr = canceled_data.groupby('reservation_status_date')[['adr']].mean()
canceled_df_adr.reset_index(inplace=True)
canceled_df_adr.sort_values('reservation_status_date', inplace=True)

not_canceled_data = df[df['is_canceled'] == 0]
not_canceled_df_adr = not_canceled_data.groupby('reservation_status_date')[['adr']].mean()
not_canceled_df_adr.reset_index(inplace=True)
not_canceled_df_adr.sort_values('reservation_status_date', inplace=True)

plt.figure(figsize=(20,6))
plt.title('Average Daily Rate')

plt.plot(
    not_canceled_df_adr['reservation_status_date'],
    not_canceled_df_adr['adr'],
    label='not canceled'
)

plt.plot(
    canceled_df_adr['reservation_status_date'],
    canceled_df_adr['adr'],
    label='canceled'
)

plt.legend()
plt.show()
```

- Longer lead times lead to more cancellations
